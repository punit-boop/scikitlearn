```python
import numpy as np
import pandas as pd
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
np.random.seed(9)
seg1= np.random.normal(loc=30,scale=5,size=100)
seg2= np.random.normal(loc=60,scale=53,size=130)
seg3= np.random.normal(loc=90,scale=15,size=90)
```

```python
data = np.concatenate([seg1,seg2,seg3]).reshape(-1,1)
data
```

```python
sc= StandardScaler()
sdata= sc.fit_transform(data)
```

```python
ncomp = 3
gmm= GaussianMixture(n_components=ncomp,random_state=9)
```

```python
gmm.fit(sdata)
```

```python
label= gmm.predict(sdata)
```

```python
plt.scatter(data,np.zeros_like(data),c= label,cmap='rainbow')
```
