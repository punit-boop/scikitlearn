```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder
```

```python
data = np.array([[25,50000,3],[35,90000,2],[40,60000,5],[45, 80000,3],
                 [24,30000,2],[55, 120000,4],[28,40000,1],[32,100000,3],
                 [38,75000,2]])
target=np.array([0,1,1,0,1,0,1,0,1])
```

```python
df= pd.DataFrame(data)
#df['target']= pd.DataFrame(target)
df
```

```python
df = df.rename(columns={
    df.columns[0]:'age',
    df.columns[1]:'salary',
    df.columns[2]:'sex',
    })
```

```python
df
```

```python
enc = OneHotEncoder()
df['sex']= enc.fit_transform(df[['sex']]).toarray()[:,1]
df
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(df,target,test_size=0.3,random_state=9)
```

```python
bst =[]
rng=[]
for x in range(10,1000,50):
    rf = RandomForestClassifier(n_estimators=x,random_state=9)
    rf.fit(xtrain,ytrain)
    pred = rf.predict(xtest)
    acc = accuracy_score(pred,ytest)
    bst.append(acc)
    rng.append(x)
```

```python
dict= {'range':rng,'accuracy':bst}
```

```python
acdf = pd.DataFrame(dict)
```
