```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
np.random.seed(42)
data = {'annual':np.random.randint(30000,150000,100),
        "spedning score": np.random.randint(1,100,100)}
data= pd.DataFrame(data)
data
```

```python
plt.scatter(data['annual'],data['spedning score'])
plt.title("Income vs SS")
plt.xlabel("Income")
plt.ylabel("Spedning Score")
```

```python
x= data.values
cluster = KMeans(n_clusters=3,random_state=9)
data['cluster']=cluster.fit_predict(x)
data['index']=np.arange(0,100,1)
data
```

```python
plt.scatter(data['annual'],data['spedning score'],c=data['cluster'],cmap='rainbow')
plt.title("Income vs SS")
plt.xlabel("Income")
plt.ylabel("Spedning Score")
```
